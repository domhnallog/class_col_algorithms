## Introduction to Statistics 
#### Calculate the mean, mode and median of `MIN` and `PTS` fields


```python
min_mean = nba_df.MIN.mean().round()
min_medi = nba_df.MIN.median()
try:
    min_mode = mode(nba_df.MIN)
except:
    min_mode = 'Error, multiple modes'

print(f"Minutes Played-   Mean: {min_mean},   Median: {min_medi},   Mode: {min_mode}")
```

```python
pts_mean = nba_df.PTS.mean().round()
pts_medi = nba_df.PTS.median()
try:
    pts_mode = mode(nba_df.PTS)
except:
    pts_mode = 'Error, multiple modes'

print(f"Points-   Mean: {pts_mean},   Median: {pts_medi},   Mode: {pts_mode}")
```

#### From the description of the data find out what information is contained in the `MIN` and `PTS` fields

### Compute the average number of minutes all the players spend on the court

```python
nba_df.MIN.mean().round()
print(f'Total Average Minutes Played per Season: {nba_df.MIN.mean().round(1)}')
```

```python
tot_avg_game_min = (nba_df.MIN.mean() / nba_df['Games Played'].mean()).round(1)
print(f'Total Average Minutes Played per Game: {tot_avg_game_min}')
```

```python
nba_df['avg_game_min'] = (nba_df.MIN / nba_df['Games Played']).round(1)
nba_df.avg_game_min.head()
```

```python
AoAp = sum(nba_df['Games Played'] * nba_df.avg_game_min) / sum(nba_df['Games Played'])
print(f"Averge of Individuals' Average Minutes per Game: {round(AoAp,1)}")
```

### Calculate the average number of minutes all teams spend on the court

```python
team_avgs = nba_df.groupby('Team').agg({'MIN':['mean','count']}).reset_index()
team_avgs
```

```python
AoAt = sum(team_avgs.MIN['count'] * team_avgs.MIN['mean']) / len(nba_df)
print(f"Averge of Teams' Average Minutes per Season: {round(AoAt,1)}")
```
